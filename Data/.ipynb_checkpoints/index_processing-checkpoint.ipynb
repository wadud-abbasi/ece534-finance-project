```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
indexes_folder = "Indexes"
djia_file = os.path.join(indexes_folder, "DJIA.csv")
nasdaq_file = os.path.join(indexes_folder, "NASDAQCOM.csv")
sp500_file = os.path.join(indexes_folder, "SP500.csv")
```

```python
def read_index_csv(file_path, value_col_name):
    df = pd.read_csv(file_path)
    df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
    df = df.rename(columns={
        'observation_date': 'date',
        df.columns[-1]: value_col_name
    })
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    print(f"\nPreview of {value_col_name} data:")
    display(df.head())
    return df[['date', value_col_name]]

djia_df = read_index_csv(djia_file, 'djia')
nasdaq_df = read_index_csv(nasdaq_file, 'nasdaqcom')
sp500_df = read_index_csv(sp500_file, 'sp500')
```

```python
merged_index_df = pd.merge(djia_df, nasdaq_df, on='date', how='outer')
print("\nAfter merging DJIA and NASDAQ:")
display(merged_index_df.head(10))
```

```python
merged_index_df = pd.merge(merged_index_df, sp500_df, on='date', how='outer')
print("\nAfter merging with SP500:")
display(merged_index_df.head(10))
```

```python
merged_index_df.to_csv("processed_indexes.csv", index=False)
```

```python
# Shape and columns
print("Merged index data shape:", merged_index_df.shape)
print("Columns:", merged_index_df.columns.tolist())

# Date range for each index (non-missing values)
for col in ['djia', 'nasdaqcom', 'sp500']:
    col_df = merged_index_df[['date', col]].dropna()
    print(f"{col.upper()} covers {col_df['date'].min().date()} to {col_df['date'].max().date()} ({len(col_df)} records)")
```

```python
# Number of missing values for each index
print("Missing values per column:")
display(merged_index_df.isnull().sum())

# Percentage of complete cases (all 3 indices present)
all_complete = merged_index_df.dropna().shape[0]
total_rows = merged_index_df.shape[0]
print(f"Rows with all indices present: {all_complete} of {total_rows} ({100*all_complete/total_rows:.2f}%)")
```

```python
# Dates where all indices are present
print("First 10 dates with all indices present:")
display(merged_index_df.dropna().head(10))

# Dates where only NASDAQ is present
print("First 10 dates with only NASDAQ present:")
display(merged_index_df[(merged_index_df['djia'].isna()) & (merged_index_df['sp500'].isna()) & (merged_index_df['nasdaqcom'].notna())].head(10))
```

```python
# Summary statistics for each index
print("Summary statistics for the indices:")
display(merged_index_df[['djia', 'nasdaqcom', 'sp500']].describe())

# Find largest daily change (simple)
for col in ['djia', 'nasdaqcom', 'sp500']:
    series = merged_index_df[col]
    max_change = series.diff().abs().max()
    print(f"Largest absolute change in {col}: {max_change}")
```

```python
subset = merged_index_df.dropna().copy()  # Only dates where all three indices available
plt.figure(figsize=(12,6))
plt.plot(subset['date'], subset['djia'], label='DJIA')
plt.plot(subset['date'], subset['nasdaqcom'], label='NASDAQ')
plt.plot(subset['date'], subset['sp500'], label='S&P500')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Major US Indices Overlapping Period')
plt.legend()
plt.show()
```
