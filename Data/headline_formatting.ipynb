```python
import pandas as pd

# Load headlines
input_path = "processed_headlines_subset.csv"  # adjust as needed
df = pd.read_csv(input_path)

# Convert "Date" to datetime and keep only date part (no time)
df['Date'] = pd.to_datetime(df['Date']).dt.date

# Group by 'Date' and 'Stock_symbol', aggregate headlines into lists
grouped = df.groupby(['Date', 'Stock_symbol'])['Article_title'].apply(list).reset_index()

# Expand list of headlines so each gets its own column: 'headline 1', 'headline 2', ...
headlines_expanded = grouped['Article_title'].apply(pd.Series)
headlines_expanded.columns = [f'headline {i+1}' for i in range(headlines_expanded.shape[1])]

# Concatenate group info with expanded headlines
result = pd.concat([grouped[['Date', 'Stock_symbol']], headlines_expanded], axis=1)

# Save to CSV
result.to_csv("processed_headlines_grouped.csv", index=False)
```
